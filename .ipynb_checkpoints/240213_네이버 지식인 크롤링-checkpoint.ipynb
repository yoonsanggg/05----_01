### 1. 네이버 지식인의 질문만 가져오기

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

html=requests.get('https://search.naver.com/search.naver?ssc=tab.kin.kqna&where=kin&sm=tab_jum&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC')
soup=bs(html.text,'html.parser')
soup
```

```python
a_tags=soup.select('a.question_text')
a_tags
```

```python
res=[i.text for i in a_tags]
res
```

### 2. 네이버 지식인 질문과 링크 가져오기

```python
res=[]
for a in a_tags:
    tmp={}
    tmp['질문']=a.text
    tmp['링크']=a['href']
    res.append(tmp)
pd.DataFrame(res)
```

### 3. 네이버 지식인 _질문, 응답, 작성자, 작성일, 링크 가져오기

```python
div_tags=soup.select('div.kin_wrap')
div_tags
```

```python
res=[]
for div in div_tags:
    tmp={}
    tmp['질문']=div.select_one('a.question_text').text
    tmp['응답']=div.select_one('a.answer_text').text
    tmp['작성자']=div.select_one('span.name').text
    tmp['작성일']=div.select('span.etc')[-1].text
    tmp['url']=div.select_one('a.question_text')['href']
    res.append(tmp)
df=pd.DataFrame(res)
df
```

### 4. 질문과 응답 세부내용 가져오기

```python
res=[]
for url in df['url']:
    html=requests.get(url)
    soup=bs(html.text,'html.parser')

    div_tags=soup.select('div.answer-content__list div.se-main-container')

    for div_tag in div_tags:
        tmp={}
        tmp['제목']=soup.select_one('div._questionContentsArea>div.c-heading__title').text.strip()
        tmp['상세질문']=soup.select_one('div._questionContentsArea>div.c-heading__content').text.strip()
        tmp['상세응답']=div_tag.text.strip()
        res.append(tmp)
    
df2=pd.DataFrame(res)
df2
```
