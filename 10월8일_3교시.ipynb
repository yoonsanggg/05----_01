```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
url='https://kin.naver.com/search/list.naver?query=%EC%A0%84%EC%84%B8%EC%82%AC%EA%B8%B0&page=1'
```

```python
data = pd.read_csv('./전세사기_네이버지식인.csv')
```

```python
data['링크'][0]
```

```python
def naver_kin(link):
    html = requests.get(link)
    soup = bs(html.text,'html')
    s1 = soup.find('div','contentArea _contentWrap')

    answer_tag = soup.find_all('div','answerDetail _endContents _endContentsText')
    res=[]
    
    for i in answer_tag:
        tmp={}
        tmp['제목']= s1.find('div','endTitleSection').text.replace('질문','').strip()
        tmp['조회수'] = s1.find_all('span','infoItem')[0].text.replace('조회수 ','').replace(',','')
        tmp['날짜'] = s1.find_all('span','infoItem')[1].text.replace('작성일','')
        tmp['질문'] = s1.find('div','questionDetail').text.strip()
        try:
            tmp['답변'] = i.find('div','se-component se-text se-l-default').text.replace('\u200b','').strip()
        except:
            tmp['답변'] = i.text
        res.append(tmp)
    df= pd.DataFrame(res)
    return df

res=[]
for i in data['링크']:
    res.append(naver_kin(i))

df = pd.concat(res,ignore_index=True)
```

```python
df.to_csv('./전세사기 세부내역_네이버지식인.csv')
```
